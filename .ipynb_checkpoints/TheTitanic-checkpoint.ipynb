```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./train.csv', index_col='PassengerId')
train_data = df[df.columns.drop(['Survived', 'Name'])]
y = df['Survived']
```

```python
train_data.info()
```

```python
train_data['Age'].fillna(np.mean(train_data['Age']), inplace=True)
train_data['Embarked'].fillna('backfill', inplace=True)
```

```python
train_data['Cabin'].unique()
```

```python
train_data['Cabin_letter'] = train_data['Cabin'].apply(lambda x: str(x)[0])
train_data['Cabin_count'] = train_data['Cabin'].apply(lambda x: 0 if pd.isna(x) else len(x.split(' ')))
```

```python
from sklearn.preprocessing import LabelEncoder

lab_enc = LabelEncoder()

cat_features = ['Sex', 'Cabin', 'Embarked', 'Ticket']

for feat in cat_features:
    train_data[feat + '_label'] = lab_enc.fit_transform(train_data[feat].astype(str))
```

```python
train_data.groupby('Ticket_label').count()
```

```python
from sklearn.ensemble import RandomForestClassifier as rf
from sklearn import metrics

def train_random_forest(X_train, X_val, y_train, y_val):
    feature_columns = X_train.columns.drop(cat_features)
    
    rand_forest = rf(random_state=0)
    rand_forest.fit(X_train[feature_columns], y_train)
    
    val_predicted = rand_forest.predict(X_val[feature_columns])
    val_roc = metrics.roc_auc_score(y_val, val_predicted)
    print('ROC score = ', val_roc)
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_data, y, random_state=0)
train_random_forest(X_train, X_val, y_train, y_val)
```

### Categorical Encoding
- Count Encoding
- Target Encoding


```python
import category_encoders as ce

count_features = ['Embarked', 'Sex']
count_enc = ce.CountEncoder(cols=count_features).fit(X_train[count_features])

train_encoded = X_train.join(count_enc.transform(X_train[count_features]).add_suffix('_count'))
test_encoded = X_val.join(count_enc.transform(X_val[count_features]).add_suffix('_count'))
```

```python
train_random_forest(train_encoded, test_encoded, y_train, y_val)
```

```python
target_features = ['Sex', 'Cabin', 'SibSp', 'Parch', 'Embarked', 'Ticket']
target_enc = ce.TargetEncoder(cols=target_features).fit(X_train[target_features], y_train)

train_encoded = X_train.join(target_enc.transform(X_train[target_features]).add_suffix('_target'))
test_encoded = X_val.join(target_enc.transform(X_val[target_features]).add_suffix('_target'))

train_random_forest(train_encoded, test_encoded, y_train, y_val)
```

### Feature Engineering
- Interactions

```python
import itertools

interactions = pd.DataFrame(index=train_data.index)
lab_enc = LabelEncoder()

cat_features = ['Age', 'Sex', 'Cabin', 'SibSp', 'Parch', 'Embarked', 'Ticket', 'Pclass']
for col1, col2 in itertools.combinations(cat_features, r=2):
    interaction = train_data[col1].map(str) + '_' + train_data[col2].map(str)
    interactions[col1+'_'+col2] = lab_enc.fit_transform(interaction)

train_data.join(interactions)

X_train, X_val, y_train, y_val = train_test_split(train_data, y, random_state=0)
train_random_forest(X_train, X_val, y_train, y_val)
```
